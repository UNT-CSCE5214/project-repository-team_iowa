**Determine which games represent about 10% of the total reviews to use as a test set**

```python
df_grouped = df.groupby(by=["title"]).count()
df_grouped["percentage"] = df_grouped["review_id"] / len(df)
df_grouped = df_grouped.sort_values(by="percentage", ascending=False)
df_grouped.head(5)
```

Taking the first two games gives us Robocraft + Eternal Card Game (9.33447%) - need another ~0.6%

```python
df_grouped = df_grouped[df_grouped["percentage"] > 0.006].sort_values(by="percentage", ascending=True)
df_grouped.head(5)
```

**Held-out test set consists of:**
* Crusaders of the Lost Idols (0.75%)
* Robocraft (4.8%)
* Eternal Card Game (4.5%)
