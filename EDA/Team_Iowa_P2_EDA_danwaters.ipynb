**Determine which game represents about 10% of the total reviews to use as a test set**

```python
df_grouped = df.groupby(by=["title"]).count()
df_grouped["percentage"] = df_grouped["review_id"] / len(df_grouped)
df_grouped = df_grouped.sort_values("percentage")
df_grouped.head(5)
```

```python
df_grouped = df_grouped[df_grouped["percentage"] > 10.0]
df_grouped.head(5)
```
